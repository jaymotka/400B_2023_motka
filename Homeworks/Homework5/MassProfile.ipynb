```python
# Load Modules
import numpy as np
import astropy.units as u
from astropy.constants import G

# import plotting modules
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline

from ReadFile import Read
from CenterOfMass import CenterOfMass
```

```python
class MassProfile:
    """
    Class to decide the mass profile of the given galaxy at the given snapshot.
    """
    
    def __init__(self, galaxy, snap):
        """
        This function initializes the class to calculate mass profile of a galaxy at the particular time using the
        specified snapshot.
        
        Parameters
        ----------
            galaxy: str
                Name of the galaxy: "MW", "M31", or "M33".
            snap: int
                Snapshot number, e.g. , 0, 1, etc.
        """
        
        # Generating the file name:
        ilbl = '000' + str(snap)       # adding a string of the filenumber to the value “000”
        ilbl = ilbl[-3:]               # removing all but the last 3 digits
        self.filename="%s_"%(galaxy) + ilbl + '.txt'   # putting together the filename
        
        self.gname = galaxy
        
        # Reading data in the given file using Read
        self.time, self.total, self.data = Read(self.filename)  # Storing data and snapshot details 
        
        # Storing the mass and positions of the particles: 
        self.m = self.data['m']               # Storing mass
        self.x = self.data['x']*u.kpc         # Storing x-coordinate
        self.y = self.data['y']*u.kpc         # Storing y-coordinate
        self.z = self.data['z']*u.kpc         # Storing z-coordinate
        
    def MassEnclosed(self, ptype, radii):
        """
        This function calculates the enclosed mass of given particle type for an array of radii and returns an array of masses.

        Parameters
        ----------
            ptype: int
                Integer representing the particle type. Use 1=Dark Matter, 2=Disk Star, 3=Bulge Star.
            radii: numpy.array
                Numpy array of radii (float) in kpc.

        Returns
        -------
            masses_enc: numpy.array
                Numpy array of particle mass (astropy.Quanity) in Msun within the corresponding radii.
        """

        gal_COM = CenterOfMass(self.filename, ptype)     # Creating CenterOfMass object
        gal_COM_p = gal_COM.COM_P(0.1)                   # Finding the COM positions by calling COM_P method with delta=0.1
        
        index = np.where(self.data["type"]==ptype)
        
        x_new = self.x[index] - gal_COM_p[0]             # Storing new x-coordinates of the particles in COM reference frame.
        y_new = self.y[index] - gal_COM_p[1]             # Storing new y-coordinates of the particles in COM reference frame.
        z_new = self.z[index] - gal_COM_p[2]             # Storing new z-coordinates of the particles in COM reference frame.
        r_new = np.sqrt(x_new**2 + y_new**2 + z_new**2)  # Storing distances of the particles in COM reference frame.
        m_new = self.m[index]
        
        masses_enc = np.zeros(len(radii))                  # Creating a numpy array with zeros to stor the enclosed masses.
        for i in range(len(radii)):                        # Looping over all the raddii.
            index2 = np.where(r_new <= radii[i]*u.kpc)     # Finding index of the particles within the given radius.
            mass_total = np.sum(m_new[index2])             # Summing up the mass of the particles within the given radius.
            masses_enc[i] = mass_total                     # Storing the enclosed mass within the given radius

        masses_enc = masses_enc*1e+10*u.Msun               # Converting to the right order of magnitude and units.
        return masses_enc
    
    def MassEnclosedTotal(self, radii):
        """
        This function calculates the total enclosed mass for an array of radii and returns an array of masses.

        Parameters
        ----------
            radii: numpy.array
                Numpy array of radii (float) in kpc.

        Returns
        -------
            mtotal_enc: numpy.array
                Numpy array of total mass (astropy.Quanity) in Msun within the corresponding radii.
        """
        
        mtotal_enc = 0                                           # Creating a variable that will store total enclosed mass
        for ptype in [1,2,3]:                                    # Looping over all three particle types
            if (self.gname == "M33") and (ptype == 3):           # Condition for bulge particles in M33
                mtotal_enc += 0                                  # Ignoring bulge particles in M33 since there are none.
            else:                                                # For all the other conditions
                m_part = self.MassEnclosed(ptype, radii)/u.Msun  # Calculating enclosed mass of the given particle type.
                mtotal_enc += m_part                             # Adding up the mass enclosed for each particle type.
        mtotal_enc = mtotal_enc*u.Msun                           # Adding the units of Msun
        
        return mtotal_enc
    
    def HernquistMass(self, r, a, Mhalo):
        """
        This function calculates the enclosed halo mass within the given radius according to Hernquist mass profile.
        
                        M(r) = (Mhalo * r**2) / (a+r)**2
        
        Parameters
        ----------
            r: float or numpy.array of floats
                Radius of enclosed halo mass in kpc.
            a: float
                Scale factor of the Hernquist mass profile in kpc.
            Mhalo: astropy.Quantity
                Total halo mass in Msun.
                
        Returns
        -------
            Mr: astropy.Quantity or numpy.array of astropy.Quantity
                The enclosed halo mass in Msun within the given radius according to Hernquist mass profile. 
        """
        
        Mr = (Mhalo * r**2) / (a+r)**2            # Enclosed Hernquist halo mass
        
        return Mr
    
    def CircularVelocity(self, ptype, radii):
        """
        This function calculates the circular speeds at given radii in km/s using the enclosed masses within the given radii 
        for a given partical type.
        
                        Vc = np.sqrt(G*Mr/radii)
        
        Parameters
        ----------
            ptype: int
                Integer representing the particle type. Use 1=Dark Matter, 2=Disk Star, 3=Bulge Star.
            radii: numpy.array
                Numpy array of radii (float) in kpc.

        Returns
        -------
            Vc: numpy.array
                Numpy array of circular speeds (astropy.Quantity) in km/s. 
        """
        
        grav = G.to(u.kpc*u.km**2/u.s**2/u.Msun)      # Converting the units of  the gravitational constant
        Mr = self.MassEnclosed(ptype, radii)          # Calculating the enclosed masses within given radii for given ptype
        Vc = np.sqrt(grav*Mr/radii/u.kpc)             # Calculating the circular speeds within given radii for given ptype 
        Vc = np.around(Vc, decimals=2)                # Rounding to 2 decimals
        
        return Vc
    
    def CircularVelocityTotal(self, radii):
        """
        This function calculates the total circular speeds for given radii in km/s using the total enclosed mass within the 
        given radii.
        
        Parameters
        ----------
            radii: numpy.array
                Numpy array of radii (float) in kpc.

        Returns
        -------
            Vcirc: numpy.array
                Numpy array of total circular speeds (astropy.Quantity) in km/s. 
        """
        
        grav = G.to(u.kpc*u.km**2/u.s**2/u.Msun)       # Converting the units of  the gravitational constant
        Mtotalr = self.MassEnclosedTotal(radii)        # Calculating the total enclosed masses within given radii 
        Vcirc = np.sqrt(grav*Mtotalr/radii/u.kpc)      # Calculating the total circular speeds within given radii
        Vcirc = np.around(Vcirc, decimals=2)           # Rounding to 2 decimals     
        
        return Vcirc  
    
    def HernquistVCirc(self, r, a, Mhalo):
        """
        This function calculates the circular speed at the given radius using the Hernquist mass profile.
        
        Parameters
        ----------
            r: float or numpy.array of floats
                Radius of enclosed halo mass in kpc.
            a: float
                Scale factor of the Hernquist mass profile in kpc.
            Mhalo: astropy.Quantity
                Total halo mass in Msun.
                
        Returns
        -------
            HernqVc: astropy.Quantity or numpy.array of astropy.Quantity
                The circular speed in km/s for the given radius according to Hernquist mass profile.
        """
        
        grav = G.to(u.kpc*u.km**2/u.s**2/u.Msun)       # Converting the units of  the gravitational constant
        Mr = self.HernquistMass(r, a, Mhalo)           # Calculating the enclosed mass within the given radius 
                                                       # using Hernquist mass profile 
        HernqVc = np.sqrt(grav*Mr/r/u.kpc)             # Calculating the total circular speeds within given radii
        HernqVc = np.around(HernqVc, decimals=2)       # Rounding to 2 decimals     
        
        return HernqVc
```

```python
def mass_profile_plotting(gal, a, r):
    """
    This function plots the different mass profiles for a given galaxy.
    
    Parameters
    ----------
        gal: str
            Name of the galaxy: "MW", "M31", or "M33".
        a: float
            Scale radius for the Hernquist profile in kpc.
        r: numpy.array
            Numpy array of radii (floats) in kpc.
    """
    
    # Calling the MassProfile object:
    gal_obj = MassProfile(gal, 0)
    
    # Calculating the total halo mass for the Hernquist profile:
    Mhalo = np.sum(gal_obj.m[np.where(gal_obj.data['type']==1)])*1e+10*u.Msun

    # Calculating the mass profiles:
    mass_dm = gal_obj.MassEnclosed(1,r)
    mass_ds = gal_obj.MassEnclosed(2,r)
    if gal != "M33": 
        mass_bs = gal_obj.MassEnclosed(3,r)
    mass_total = gal_obj.MassEnclosedTotal(r)
    mass_Hernq = gal_obj.HernquistMass(r, a, Mhalo)

    # Plotting:
    fig = plt.figure(figsize=(10,10))
    ax = plt.subplot(111)

    # Plotting the mass profiles:
    plt.plot(r, mass_dm, color="black", linestyle='--', label="Dark Matter")
    plt.plot(r, mass_ds, color="blue", linestyle='--', label="Disk Stars")
    if gal != "M33":
        plt.plot(r, mass_bs, color="red", linestyle='--', label="Bulge Stars")
    plt.plot(r, mass_total, color="green", label="Total Mass")
    plt.plot(r, mass_Hernq, color="orange", linestyle='-.', label="Hernquist Mass")

    # Coverting to log-scale:
    plt.yscale("log")

    
    # Adding Title:
    plt.title("Mass Profile")
    
    # Adding axis labels
    plt.xlabel('Radius [kpc]', fontsize=22)
    plt.ylabel(r'M(R) [M$_\odot$]', fontsize=22)

    # Adjusting tick label font size
    label_size = 22
    mpl.rcParams['xtick.labelsize'] = label_size 
    mpl.rcParams['ytick.labelsize'] = label_size

    # Adding a legend with some customizations
    legend = ax.legend(loc='lower right',fontsize='x-large')

    # Adding figure text
    plt.figtext(0.15, 0.83, gal, fontsize=22)

    plt.savefig(gal + '_massprofile.png')
```

```python
def rotation_curve_plotting(gal, a, r):
    """
    This function plots the different mass profiles for a given galaxy.
    
    Parameters
    ----------
        gal: str
            Name of the galaxy: "MW", "M31", or "M33".
        a: float
            Scale radius for the Hernquist profile in kpc.
        r: numpy.array
            Numpy array of radii (floats) in kpc.
    """
    
    # Calling the MassProfile object:
    gal_obj = MassProfile(gal, 0)
    
    # Calculating the total halo mass for the Hernquist profile:
    Mhalo = np.sum(gal_obj.m[np.where(gal_obj.data['type']==1)])*1e+10*u.Msun

    # Calculating the mass profiles:
    vc_dm = gal_obj.CircularVelocity(1,r)
    vc_ds = gal_obj.CircularVelocity(2,r)
    if gal != "M33": 
        vc_bs = gal_obj.CircularVelocity(3,r)
    vc_total = gal_obj.CircularVelocityTotal(r)
    vc_Hernq = gal_obj.HernquistVCirc(r, a, Mhalo)

    # Plotting:
    fig = plt.figure(figsize=(10,10))
    ax = plt.subplot(111)

    # Plotting the mass profiles:
    plt.plot(r, vc_dm, color="black", linestyle='--', label="Dark Matter")
    plt.plot(r, vc_ds, color="blue", linestyle='--', label="Disk Stars")
    if gal != "M33":
        plt.plot(r, vc_bs, color="red", linestyle='--', label="Bulge Stars")
    plt.plot(r, vc_total, color="green", label="Total VCirc")
    plt.plot(r, vc_Hernq, color="orange", linestyle='-.', label="Hernquist VCirc")

    # Adding Title:
    plt.title("Rotation Curve")
    
    # Adding axis labels
    plt.xlabel('Radius [kpc]', fontsize=22)
    plt.ylabel('V_Circ [km/s]', fontsize=22)

    # Adjusting tick label font size
    label_size = 22
    mpl.rcParams['xtick.labelsize'] = label_size 
    mpl.rcParams['ytick.labelsize'] = label_size

    # Adding a legend with some customizations
    legend = ax.legend(loc='lower right',fontsize='x-large')

    # Adding figure text
    plt.figtext(0.8, 0.83, gal, fontsize=22)

    plt.savefig(gal + '_rotationcurve.png')
```
