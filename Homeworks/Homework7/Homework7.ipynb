# M33AnalyticOrbit

```python
class M33AnalyticOrbit:
    """ Calculates the analytical orbit of M33 around M31 """
    
    def __init__(self, output_file): # **** add inputs
        ''' Class to calculate thee the acceleration M33 feels from M31 and 
        integrate its current position and velocity forwards in time. 
            
            PARAMETERS
            ----------
            output_file : `str`
                filename to store the data of the integrated orbit
        '''

        ### get the gravitational constant (the value is 4.498502151575286e-06)
        self.G = const.G.to(u.kpc**3/u.Msun/u.Gyr**2).value
        
        ### **** store the output file name
        self.filename = output_file
        
        ### get the current pos/vel of M33 
        # **** create an instance of the  CenterOfMass class for M33
        M33_COM = CenterOfMass('M33_000.txt', 2)

        # **** store the position VECTOR of the M33 COM (.value to get rid of units)
        self.M33_COM_p = M33_COM.COM_P(0.1, 4).value
    
        # **** store the velocity VECTOR of the M33 COM (.value to get rid of units)
        self.M33_COM_v = M33_COM.COM_V(self.M33_COM_p[0]*u.kpc, self.M33_COM_p[1]*u.kpc, self.M33_COM_p[2]*u.kpc).value
        
        ### get the current pos/vel of M31 
        # **** create an instance of the  CenterOfMass class for M31 
        M31_COM = CenterOfMass('M31_000.txt', 2)

        # **** store the position VECTOR of the M31 COM (.value to get rid of units)
        self.M31_COM_p = M31_COM.COM_P(0.1, 2).value
    
        # **** store the velocity VECTOR of the M31 COM (.value to get rid of units)
        self.M31_COM_v = M31_COM.COM_V(self.M31_COM_p[0]*u.kpc, self.M31_COM_p[1]*u.kpc, self.M31_COM_p[2]*u.kpc).value
        
        ### store the DIFFERENCE between the vectors posM33 - posM31
        # **** create two VECTORs self.r0 and self.v0 and have them be the
        # relative position and velocity VECTORS of M33
        self.r0 = self.M33_COM_p - self.M31_COM_p
        self.v0 = self.M33_COM_v - self.M31_COM_v
        
        ### get the mass of each component in M31 
        ### disk
        # **** self.rdisk = scale length (no units)
        self.rdisk = 5

        # **** self.Mdisk set with ComponentMass function. Remember to *1e12 to get the right units. Use the right ptype
        self.Mdisk = ComponentMass('M31_000.txt', 2).value*1e12
        
        ### bulge
        # **** self.rbulge = set scale length (no units)
        self.rbulge = 1

        # **** self.Mbulge  set with ComponentMass function. Remember to *1e12 to get the right units Use the right ptype
        self.Mbulge = ComponentMass('M31_000.txt', 3).value*1e12
        
        # Halo
        # **** self.rhalo = set scale length from HW5 (no units)
        self.rhalo = 60

        # **** self.Mhalo set with ComponentMass function. Remember to *1e12 to get the right units. Use the right ptype
        self.Mhalo = ComponentMass('M31_000.txt', 1).value*1e12
    
    def HernquistAccel(self, M, ra, r): # it is easiest if you take as an input the position VECTOR 
        ''' Method to compute the gravitational acceleration induced by a Hernquist profile for halo or bulge.
        
        PARAMETERS
        ----------
        M : float
            total mass of the component
        ra : float
            scale radius of the component
        r : numpy.array of floats
            numpy array containing position coordinates
        
        RETURNS
        -------
        Hern: numpy.array of floats
            numpy array containing acceleration coordinates
        '''
        
        ### **** Store the magnitude of the position vector
        rmag = np.sqrt(r[0]**2 + r[1]**2 + r[2]**2)
        
        ### *** Store the Acceleration
        #follow the formula in the HW instructions
        Hern = ( -self.G*M/(rmag *(ra + rmag)**2) ) * r  
        
        # NOTE: we want an acceleration VECTOR so you need to make sure that in the Hernquist equation you 
        # use  -G*M/(rmag *(ra + rmag)**2) * r --> where the last r is a VECTOR 
        
        return Hern
    
    def MiyamotoNagaiAccel(self, M, rd, r):# it is easiest if you take as an input a position VECTOR  r 
        ''' Method to compute the gravitational acceleration induced by Miyamoto-Nagai 1975 profile for disk.
        
        PARAMETERS
        ----------
        M : float
            total mass of the disk
        ra : float
            scale radius of the disk
        r : numpy.array of floats
            numpy array containing position coordinates
        
        RETURNS
        -------
        disk_acc: numpy.array of floats
            numpy array containing acceleration coordinates
        '''
        
        ### Acceleration **** follow the formula in the HW instructions
        # AGAIN note that we want a VECTOR to be returned  (see Hernquist instructions)
        # this can be tricky given that the z component is different than in the x or y directions. 
        # we can deal with this by multiplying the whole thing by an extra array that accounts for the 
        # differences in the z direction:
        #  multiply the whle thing by :   np.array([1,1,ZSTUFF]) 
        # where ZSTUFF are the terms associated with the z direction
        
        R = np.sqrt(r[0]**2 + r[1]**2)
        zd = rd/5
        B = rd + np.sqrt(r[2]**2 + zd**2)
        extra_arr = np.array([1,1, B/np.sqrt(r[2]**2 + zd**2)])
        
        disk_acc = ( -self.G*M/(R**2+B**2)**1.5 ) * r * extra_arr
        
        return disk_acc 
        # the np.array allows for a different value for the z component of the acceleration
     
    
    def M31Accel(self, r): # input should include the position vector, r
        ''' Method to compute the total gravitational acceleration by summing up the acceleration 
        due to each components. 
        
        PARAMETERS
        ----------
        r : numpy.array of floats
            numpy array containing position coordinates
        
        RETURNS
        -------
        net_acc: numpy.array of floats
            numpy array containing total acceleration coordinates
        '''
        
        ### Call the previous functions for the halo, bulge and disk
        # **** these functions will take as inputs variable we defined in the initialization of the class like 
        # self.rdisk etc. 
        
        halo_acc = self.HernquistAccel(self.Mhalo, self.rhalo, r)
        bulge_acc = self.HernquistAccel(self.Mbulge, self.rbulge, r)
        disk_acc = self.MiyamotoNagaiAccel(self.Mdisk, self.rdisk, r)
        
        net_acc = halo_acc + bulge_acc + disk_acc
        
            # return the SUM of the output of the acceleration functions - this will return a VECTOR 
        return net_acc
    
    def LeapFrog(self, dt, r, v): # take as input r and v, which are VECTORS. Assume it is ONE vector at a time
        ''' Method that defines an integrator that uses a variant of Leap Frog integration scheme
        to solve the orbit of M33. 
        
        PARAMETERS
        ----------
        dt: float
            time interval for integration
            
        r : numpy.array of floats
            numpy array containing position coordinates of M33 COM relative to M31 COM at given time t_n
        
        v: numpy.array of floats
            numpy array containing velocity components of M33 COM relative to M31 COM at given time t_n
        
        RETURNS
        -------
        rnew : numpy.array of floats
            numpy array containing the new position coordinates of M33 COM relative to M31 COM at time t_n+1 = t_n + dt
        
        vnew: numpy.array of floats
            numpy array containing the new velocity components of M33 COM relative to M31 COM at time t_n+1 = t_n + dt
        '''
        
        # predict the position at the next half timestep
        rhalf = r + (v * dt / 2)
        
        # predict the final velocity at the next timestep using the acceleration field at the rhalf position 
        ahalf = self.M31Accel(rhalf)
        vnew = v + (ahalf * dt)
        
        # predict the final position using the average of the current velocity and the final velocity
        # this accounts for the fact that we don't know how the speed changes from the current timestep to the 
        # next, so we approximate it using the average expected speed over the time interval dt. 
        rnew = rhalf + (vnew * dt / 2)
        
        return rnew, vnew      # **** return the new position and velcoity vectors
    
    def OrbitIntegration(self, t0, dt, tmax):
        ''' Method that loops over the Leap Frog integrator to compute the orbit of M33 and saves the data
        in the ouput file.
        
        PARAMETERS
        ----------
        t0: float
            the starting time of the integration in Gyrs
            fixed: t0 = 0 (since we are reading the COM data for current time)  
        
        dt: float
            time interval for integration
            
        tmax :float
            the ending time of the integration in Gyrs
        '''

        # initialize the time to the input starting time
        t = t0
        
        # initialize an empty array of size :  rows int(tmax/dt)+2  , columns 7
        orbit = np.zeros([int(tmax/dt)+2, 7])
        
        # initialize the first row of the orbit
        orbit[0] = t0, *tuple(self.r0), *tuple(self.v0)
        # this above is equivalent to 
        # orbit[0] = t0, self.r0[0], self.r0[1], self.r0[2], self.v0[0], self.v0[1], self.v0[2]
    
        # initialize a counter for the orbit.  
        i = 1 # since we already set the 0th values, we start the counter at 1
        
        # initializing the position and velocity vectors that will get updated 
        # by Leap Frog integrator at each loop
        r, v = self.r0, self.v0
        
        # start the integration (advancing in time steps and computing LeapFrog at each step)
        while t < tmax:          # as long as t has not exceeded the maximal time 
            # **** advance the time by one timestep, dt
            t += dt
            
            # **** store the new time in the first column of the ith row
            orbit[i, 0] = t

            # ***** advance the position and velocity using the LeapFrog scheme
            # remember that LeapFrog returns a position vector and a velocity vector  
            # as an example, if a function returns three vectors you would call the function and store 
            # the variable like:     a,b,c = function(input)
            r, v = self.LeapFrog(dt, r, v)
            
            # ****  store the new position vector into the columns with indexes 1,2,3 of the ith row of orbit
            # TIP:  if you want columns 5-7 of the Nth row of an array called A, you would write : 
            # A[n, 5:8] 
            # where the syntax is row n, start at column 5 and end BEFORE column 8
            orbit[i,1:] = *tuple(r), *tuple(v)
            
            # **** update counter i , where i is keeping track of the number of rows (i.e. the number of time steps)
            i += 1
        
        # write the data to a file
        np.savetxt(self.filename, orbit, fmt = "%11.3f"*7, comments='#', 
                   header="{:>10s}{:>11s}{:>11s}{:>11s}{:>11s}{:>11s}{:>11s}"\
                   .format('t', 'x', 'y', 'z', 'vx', 'vy', 'vz'))
        
        # there is no return function
```

```python
# Finding orbital seperation and relative speed of M33 relative to M31 from the analytic orbital calculation:

# Loading the saved data from the analytic orbital calc
analytic_orbit = np.loadtxt('M33_analytic_orbit.txt')

# Saving time, seperation, and rel speed for analytic method:
t_analytic = analytic_orbit[:,0]
r_analytic = np.sqrt(analytic_orbit[:,1]**2 + analytic_orbit[:,2]**2 + analytic_orbit[:,3]**2)
v_analytic = np.sqrt(analytic_orbit[:,4]**2 + analytic_orbit[:,5]**2 + analytic_orbit[:,6]**2)
```

```python
# Finding orbital seperation and relative speed of M33 relative to M31 from the analytic orbital calculation:

# Loading the simulated data for M33 and M31 orbits:
M33_sim_orbit = np.loadtxt('Orbit_M33.txt')
M31_sim_orbit = np.loadtxt('Orbit_M31.txt')

# Saving time, seperation, and rel speed from the simulated data:
t_sim = M33_sim_orbit[:,0]
r_sim_vec = M33_sim_orbit[:,1:4] - M31_sim_orbit[:,1:4]
v_sim_vec = M33_sim_orbit[:,4:] - M31_sim_orbit[:,4:]
r_sim = np.sqrt(r_sim_vec[:,0]**2 + r_sim_vec[:,1]**2 + r_sim_vec[:,2]**2)
v_sim = np.sqrt(v_sim_vec[:,0]**2 + v_sim_vec[:,1]**2 + v_sim_vec[:,2]**2)
```

```python
# Plotting orbital seperation of M33 relative to M31
plt.plot(t_analytic, r_analytic, label="Analytic Orbit")
plt.plot(t_sim, r_sim, label="Simulated Orbit")
plt.title("Orbital Seperation of M33 relative to M31")
plt.xlabel("Time [Gyrs]")
plt.ylabel("Separation [kpc]")
plt.legend()
```

```python
# Plotting Relative Speed of M33 relative to M31
plt.plot(t_analytic, v_analytic, label="Analytic Orbit")
plt.plot(t_sim, v_sim, label="Simulated Orbit")
plt.title("Relative Speed of M33 relative to M31")
plt.xlabel("Time [Gyrs]")
plt.ylabel("Relative Speed [kpc]")
plt.legend()
```
