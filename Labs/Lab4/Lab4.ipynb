```python
# Some Notes about the Isochrone Data
# DATA From   http://stellar.dartmouth.edu/models/isolf_new.html
# files have been modified from download.  ( M/Mo --> M;   Log L/Lo --> L)
# removed #'s from all lines except column heading
# NOTE SETTINGS USED:  Y = 0.245 default   [Fe/H] = -2.0  alpha/Fe = -0.2
# These could all be changed and it would generate a different isochrone
```

```python

# Filename for data with Isochrone fit for different ages
# These files are located in the folder IsochroneData
filename1="./IsochroneData/Isochrone3.txt"          # Filename for Isochrone of 3 Gyr according to Tolstoy plot
filename2="./IsochroneData/Isochrone7.txt"          # Filename for Isochrone of 7 Gyr according to Tolstoy plot
filename3="./IsochroneData/Isochrone13.txt"         # Filename for Isochrone of 13 Gyr according to Tolstoy plot
filename4="./IsochroneData/Isochrone1.txt"          # Filename for Isochrone of 1 Gyr (younger isochrone than Tolstoy plot)
```

```python
# READ IN DATA
# "dtype=None" means line is split using white spaces
# "skip_header=8"  skipping the first 8 lines 
# the flag "names=True" creates arrays to store the date
#       with the column headers given in line 8 

# Read in data for an isochrone corresponding to different ages
data1 = np.genfromtxt(filename1,dtype=None,names=True,skip_header=8)   #data for isochrone of 3 Gyr according to Tolstoy plot
data2 = np.genfromtxt(filename2,dtype=None,names=True,skip_header=8)   #data for isochrone of 7 Gyr according to Tolstoy plot
data3 = np.genfromtxt(filename3,dtype=None,names=True,skip_header=8)   #data for isochrone of 13 Gyr according to Tolstoy plot
data4 = np.genfromtxt(filename4,dtype=None,names=True,skip_header=8)   #data for isochrone of 1 Gyr
```

```python
# Plot Isochrones 
# For Carina

fig = plt.figure(figsize=(10,10))
ax = plt.subplot(111)

# Plot Isochrones

# Isochrone for different ages
# Plotting Color vs. Difference in Color 
plt.plot(data1['B']-data1['R'], data1['R'], color='blue', linewidth=5, label='3 Gyr, Tolstoy') #plotting 3 Gyr isochrone
                                                                                               #suggested in Tolstoy plot

plt.plot(data2['B']-data2['R'], data2['R'], color='green', linewidth=5, label='7 Gyr, Tolstoy')#plotting 7 Gyr isochrone
                                                                                               #suggested in Tolstoy plot

plt.plot(data3['B']-data3['R'], data3['R'], color='black', linewidth=5, label='13 Gyr, Tolstoy')#plotting 13 Gyr isochrone
                                                                                               #suggested in Tolstoy plot

plt.plot(data4['B']-data4['R'], data4['R'], color='red', linewidth=5, label='1 Gyr, younger')#plotting 1 Gyr isochrone, younger
                                                                                   #than what is suggested in Tolstoy plot
###EDIT Here, following the same format as the line above 

# Add axis labels
plt.xlabel('B-R', fontsize=22)
plt.ylabel('M$_R$', fontsize=22)

#set axis limits
plt.xlim(-0.5,2)
plt.ylim(5,-2.5)

#adjust tick label font size
label_size = 22
matplotlib.rcParams['xtick.labelsize'] = label_size 
matplotlib.rcParams['ytick.labelsize'] = label_size

# add a legend with some customizations.
legend = ax.legend(loc='upper left',fontsize='x-large')
#add figure text
plt.figtext(0.6, 0.15, 'Isochrone Carina', fontsize=22)

plt.savefig('IsochroneLab4.png')
```
